# Benchmarks Results


```python
import matplotlib.pyplot as plt
import math
from matplotlib.lines import Line2D
```

```python
with open('gpu2070s_cka_benchmark.csv', 'r') as file:
    data = file.read().splitlines()

cka_new = [float(x) for x in data[0].split(',')]
cka_old = [float(x) for x in data[1].split(',')]
x = [2 ** n for n in range(len(cka_new))]
x.reverse()
```

## Non Log Plot

```python
plt.plot(x, cka_new, label="New Method")
plt.plot(x, cka_old, label="Old Method")
plt.title("C.K.A. Method Time Comparision")
plt.xlabel("Matrix Size")
plt.ylabel("Time (s)")
plt.legend()
plt.axvline(x=12720, color='gray', linestyle='--')
# create custom legend handles and labels
handles, labels = plt.gca().get_legend_handles_labels()
handles.append(Line2D([0], [0], color='gray', linestyle='--'))
labels.append('Our Model')

# add legend to plot
plt.legend(handles=handles, labels=labels)
plt.show()
```

## Log Plot

```python
plt.xscale("log", base=2)

plt.plot(x, cka_new, label="New Method")
plt.plot(x, cka_old, label="Old Method")
plt.title("C.K.A. Method Time Comparision")
plt.xlabel("Matrix Size")
plt.xscale("log", base=2)
plt.ylabel("Log Time (s)")
plt.yscale("log", base=2)
plt.legend()
plt.axvline(x=12720, color='gray', linestyle='--')

# create custom legend handles and labels
handles, labels = plt.gca().get_legend_handles_labels()
handles.append(Line2D([0], [0], color='gray', linestyle='--'))
labels.append('Our Model')

# add legend to plot
plt.legend(handles=handles, labels=labels)

plt.show()
```
