# Differing Learning Rates Notebooks

```python
import numpy as np
import pandas as pd
import json
```

Reading the data.

```python
# Creating an array for each of the files names.
files_list = np.empty((12,), dtype='U64')
files_list[0] = "s01.json"
for i in range (1, 12):
    files_list[i] = "s01" + "(1)" * i + ".json"
```

```python
# Allocating matrices for each of the measures.
hyper_parameters = np.empty((12, 7), dtype='U16')
cka = np.empty((12, 3000), dtype='double')
loss = np.empty((12, 3000), dtype='double')
accuracy = np.empty((12, 3000), dtype='double')
```

```python
# Loading the data into the matrices.
i = 0
for file in files_list:
    f = open(file)
    data = json.load(f)
    
    hyper_parameters[i] = data["Hyper Parameters"]
    cka[i] = data["Centered Kernel Alignment"]
    loss[i] = data["Loss"]
    accuracy[i] = data["Validation Accuracy"]
    i += 1
i = 0
```

```python
def smooth(y):
    box = np.ones(y.size()) / y.size()
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

Loss over Epochs

```python
nrow = 3
ncol = 4
fig, axes = plt.subplots(nrow, ncol)
title = hyper_parameters[ :,6]
for i, ax in enumerate(axes.ravel()):
    ax.set_tittle(f"Slow Learning Rate -> {title[i]}")
    ax.set_ylabel('Loss')
    ax.set_xlabel('Epochs'
    loss[i].plot[ax=ax]
```
